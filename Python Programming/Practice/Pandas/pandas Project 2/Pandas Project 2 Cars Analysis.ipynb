# Now Analyze What are the top Brands

```python
cars["brand"]
```

```python
selling_prop = cars["brand"].value_counts(normalize= True)
```

```python
selling_prop
```

```python
selling_bol = selling_prop > 0.05
```

```python
selling_bol
```

```python
top_brand = selling_prop[selling_bol]
top_brand
```

```python
top_brand.values
```

```python
top_brand.index
```

```python
brand_name = top_brand.index
```

```python
brand_bol = cars['brand'].isin(brand_name)
```

```python
top_selling_brand = cars[brand_bol].copy()
```

```python
top_selling_brand
```

```python
top_selling_brand["price"]
```

```python
top_selling_brand.rename({"price":"price ($)", "odometer":"odometer (km)"}, axis=1, inplace=True)
```

```python
top_selling_brand
```

# Know GroupBY Top_Selling_Brands

```python
groupping = top_selling_brand.groupby("brand")
```

```python
brand_mean = groupping[["price ($)","odometer (km)"]].mean()
```

```python
brand_mean
```
