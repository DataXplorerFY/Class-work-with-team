# Let's Analyze the column one-by-one

The below column data type is object we have to change it into date time for furthur operations

```python
cars["dateCrawled"]
```

```python
cars["dateCrawled"] = pd.to_datetime(cars["dateCrawled"])
```

```python
cars["price"]
```

```python
cars["price"] = cars["price"].str.replace("$","").str.replace(",","").astype(int)
```

```python
cars["price"]
```

```python
cars["odometer"] = cars["odometer"].str.replace(",","").str.replace("km","").astype(int)
```

```python
cars["odometer"]
```

```python
cars["dateCreated"] = pd.to_datetime(cars["dateCreated"])
```

```python
cars["dateCreated"]
```

```python
cars["lastSeen"] = pd.to_datetime(cars["lastSeen"])
```
