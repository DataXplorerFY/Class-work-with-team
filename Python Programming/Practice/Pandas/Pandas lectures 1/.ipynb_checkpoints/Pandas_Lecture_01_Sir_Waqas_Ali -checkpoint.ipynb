```python
bad_char = ["(",")","c","C",".",",","s"," "]
```

```python
len(bad_char)
```

```python
date = '(c. 1955-1959)'

for i in bad_char:
    date=date.replace(i," ")
    
print(date)
```

```python
def bad_char_remove(date):
    bad_char = ["(",")","c","C",".",",","s","", "'"]
    for i in bad_char:
        date=date.replace(i,"")
    return date
        
    
    
bad_char_remove('(c.c c 1955-1659)')
    
    
```

```python
arts["Date"] = arts["Date"].apply(bad_char_remove)
```

# Our Next Task Is To Clean The Ranges In The Date Column

```python
new_date = "1889-1911"
```

```python
lst = new_date.split("-")
yr1 = lst[0]
yr2 = lst[1]
yr1 = int(yr1)
yr2 = int(yr2)
yr = round((yr1 + yr2)/2)
print(yr)
```

```python
lst
```

```python
def int_date(date):
    if "-" in date:
        year_range = date.split("-")
        year1 = int(year_range[0])
        year2 = int(year_range[1])
        average_year = round((year1 + year2) / 2)
        return average_year
    else:
        return int(date)

print(int_date("1968-1969"))
```

```python
arts['Date'] = arts['Date'].apply(int_date)
```
