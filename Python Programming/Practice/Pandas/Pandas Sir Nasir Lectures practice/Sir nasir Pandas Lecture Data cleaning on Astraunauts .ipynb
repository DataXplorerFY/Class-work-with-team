# Next Part - Handling Missing values and Duplicate Records

#  we know that we do have some missing values. Let's see how many we have in each column.

```python
missing_values_count = data.isnull().sum()
```

```python
missing_values_count
```

That seems like a lot! It might be helpful to see what percentage of the values in our dataset were missing to give us a better sense of the scale of this problem

```python
# how many total missing values do we have?
total_cells = np.product(data.shape)
total_missing = missing_values_count.sum()

# percent of data that is missing
percent_missing = (total_missing/total_cells) * 100
print(percent_missing)
```

# Drop missing values

```python
# remove all the rows that contain a missing value
data.dropna()
```

```python
# remove all columns with at least one missing value
columns_with_na_dropped = data.dropna(axis=1)
columns_with_na_dropped.head()
```

```python
# just how much data did we lose?
print("Columns in original dataset: %d \n" % data.shape[1])
print("Columns with na's dropped: %d" % columns_with_na_dropped.shape[1])
```

We have lost quite a bit of data, but at this point we have successfully removed all the NaN's from our data.
